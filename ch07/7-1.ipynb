### 7.1 K-means

```python
%matplotlib inline
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_data = 1000
seed = 1
n_centers = 4
```

```python
# 4つの二変量正規分布に従うデータを生成し、k-meansを実行する
blobs, blob_labels = make_blobs(n_samples=n_data, 
                                n_features=2, 
                                centers=n_centers, 
                                random_state=seed
                                )
clusters_blob = KMeans(n_clusters=n_centers, random_state=seed).fit_predict(blobs)
```

```python
#　2次元の一様分布に従うデータを生成し、k-meansを実行する
uniform = np.random.rand(n_data, 2)
clusters_uniform = KMeans(n_clusters=n_centers, random_state=seed).fit_predict(uniform)
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))

figure, axs = plt.subplots(2, 2, figsize=(10, 10))
sns.scatterplot(x=blobs[:, 0], y=blobs[:, 1], 
                hue=blob_labels, 
                ax=axs[0, 0], 
                palette='gist_rainbow',
                alpha=0.7
                )
axs[0, 0].set_title('(a) Four randomly generated blobs', fontsize=14)
axs[0, 0].axis('off')

sns.scatterplot(x=blobs[:, 0], y=blobs[:, 1], 
                hue=clusters_blob, 
                ax=axs[0, 1], 
                palette='gist_rainbow',
                alpha=0.7
                )
axs[0, 1].set_title('(b) Clusters found via K-means', fontsize=14)
axs[0, 1].axis('off')

sns.scatterplot(x=uniform[:, 0], y=uniform[:, 1], 
                ax=axs[1, 0]
                )
axs[1, 0].set_title('(c) 1000 randomly generated points', fontsize=14)
axs[1, 0].axis('off')

sns.scatterplot(x=uniform[:, 0], y=uniform[:, 1], 
                hue=clusters_uniform,
                ax=axs[1, 1], 
                palette='gist_rainbow',
                alpha=0.7
                )
axs[1, 1].set_title('(d) Clusters found via K-means', fontsize=14)
axs[1, 1].axis('off')

plt.show()
```
